```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={"Data":pd.to_datetime(["2022-01-02","2024-05-08","2022-01-04","2022-01-02","2021-01-02"]),
      "Value":[10,15,21,29,30]}
df=pd.DataFrame(data)
df.set_index("Data",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker='x')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Trend")
plt.grid(True)
plt.show()
```

```python
data={"Data":pd.to_datetime(["2022-01-02","2024-05-08","2022-01-04","2022-01-02","2021-01-02"]),
      "Value":[19,15,21,67,30]}
df=pd.DataFrame(data)
df.set_index("Data",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker='x')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Seasonality")
plt.grid(True)
plt.show()
```
